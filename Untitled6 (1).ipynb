```python
import pandas as pd
import numpy as np
from faker import Faker

fake = Faker()
```

```python
# Number of records
n = 100000

# Generate data
data = {
    "Customer_ID": [fake.unique.random_int(min=100000, max=999999) for _ in range(n)],
    "Age": np.random.randint(18, 71, n),
    "Gender": np.random.choice(["Male", "Female"], n),
    "Tenure": np.random.randint(0, 11, n),
    "Balance": np.round(np.random.uniform(0, 200000, n), 2),
    "CreditScore": np.random.randint(300, 901, n),
    "EstimatedSalary": np.round(np.random.uniform(10000, 150000, n), 2),
    "NumOfProducts": np.random.randint(1, 5, n),
    "IsActiveMember": np.random.choice(["Yes", "No"], n),
    "Churn": np.random.choice([0, 1], n)
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("customer_data.csv", index=False)

print("Dataset generated successfully!")
print(df)
```

```python

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Verify data types
print("\nData types per column:")
print(df.info())


# Calculate descriptive statistics for numerical columns
print("Descriptive Statistics for Numerical Columns:")
display(df.describe())

# Calculate descriptive statistics for all columns (including categorical)
print("\nDescriptive Statistics for All Columns (including categorical):")
display(df.describe(include='all'))

# For mode, as df.describe(include='all') only shows the top mode, let's explicitly get the mode for all columns
print("\nMode for all columns:")
for column in df.columns:
    print(f"Column '{column}': {df[column].mode().tolist()}")




Churn Behavior Analysis
Let's start by understanding the overall churn distribution and then compare churned vs. non-churned customers based on key features.


import matplotlib.pyplot as plt
import seaborn as sns

# Overall Churn Distribution
churn_counts = df['Churn'].value_counts().reset_index()
churn_counts.columns = ['Churn', 'Count']
churn_counts['Percentage'] = (churn_counts['Count'] / len(df)) * 100

plt.figure(figsize=(6, 4))
sns.barplot(x='Churn', y='Percentage', data=churn_counts, palette='viridis')
plt.title('Overall Churn Distribution')
plt.xlabel('Churn (0 = No Churn, 1 = Churn)')
plt.ylabel('Percentage')
plt.xticks(ticks=[0, 1], labels=['No Churn', 'Churn'])
plt.show()

print(churn_counts)


# Churn by Gender
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Gender', hue='Churn', palette='pastel')
plt.title('Churn by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

# Churn by IsActiveMember
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='IsActiveMember', hue='Churn', palette='pastel')
plt.title('Churn by Active Member Status')
plt.xlabel('Is Active Member')
plt.ylabel('Number of Customers')
plt.show()

# Churn by NumOfProducts
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='NumOfProducts', hue='Churn', palette='pastel')
plt.title('Churn by Number of Products')
plt.xlabel('Number of Products')
plt.ylabel('Number of Customers')
plt.show()


# Compare numerical features for churned vs. non-churned customers
numerical_cols = ['Age', 'Tenure', 'Balance', 'CreditScore', 'EstimatedSalary']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(data=df, x='Churn', y=col, palette='coolwarm')
    plt.title(f'{col} vs. Churn')
    plt.xlabel('Churn (0 = No Churn, 1 = Churn)')
    plt.ylabel(col)
plt.tight_layout()
plt.show()


# Define age bins and labels
age_bins = [0, 25, 35, 45, 55, 65, 75, 100]
age_labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65-74', '75+']

# Create a new 'AgeGroup' column
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Calculate churn rate by AgeGroup
age_churn = df.groupby('AgeGroup')['Churn'].value_counts(normalize=True).unstack().reset_index()
if 1 in age_churn.columns:
    age_churn['ChurnRate'] = age_churn[1] * 100
else:
    age_churn['ChurnRate'] = 0 # No churn in any age group

# Plot Churn Rate by Age Group
plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='ChurnRate', data=age_churn, palette='viridis')
plt.title('Churn Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Churn Rate (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Pie Chart for Gender Distribution
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Histogram for Age Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=10, kde=True, color='purple')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# Scatter Plot: Balance vs. EstimatedSalary, colored by Churn
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='Balance', y='EstimatedSalary', hue='Churn', palette='coolwarm', alpha=0.6)
plt.title('Balance vs. Estimated Salary by Churn')
plt.xlabel('Balance')
plt.ylabel('Estimated Salary')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()


# Heatmap of Correlation Matrix for Numerical Features
numerical_df = df[['Age', 'Tenure', 'Balance', 'CreditScore', 'EstimatedSalary', 'NumOfProducts', 'Churn']]
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
